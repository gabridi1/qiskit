```python
from qiskit import *
from qiskit.tools.monitor import job_monitor
from qiskit.tools.visualization import plot_histogram
from qiskit import Aer
from qiskit import QuantumCircuit, QuantumRegister
import numpy as np
```

```python
#Exemplo
Q = QuantumRegister(5)
C = ClassicalRegister(5)
circuito = QuantumCircuit(Q, C)
circuito.x(Q[0])
circuito.h(Q)
circuito.barrier()
circuito.cx(Q[4], Q[0])
circuito.cx(Q[2], Q[0])
circuito.cx(Q[1], Q[0])
circuito.barrier()
circuito.h(Q)
circuito.barrier()
circuito.measure(Q[1], C[1])
circuito.measure(Q[2], C[2])
circuito.measure(Q[3], C[3])
circuito.measure(Q[4], C[4])

circuito.draw(output='mpl')
```

```python
simulador = BasicAer.get_backend('qasm_simulator')
job = execute(circuito, simulador, shots=8024)
resultado = job.result()
contagem = resultado.get_counts()
plot_histogram(contagem)
```
